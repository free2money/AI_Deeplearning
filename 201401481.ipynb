```python
def derivative(function, point):
    h=0.00000000000001
    return (function(point+h)-function(point))/h
```

```python
def f1(x):
    return x**2

result = f1(2)
print(result)
print(derivative(f1,2))
```

```python
def f2(x):
    return x**3+x**2+x

print(f2(2))
print(derivative(f2,2))
```

```python
from math import *
def f3(x):
    return x*sin(x*x*x+3)
print(f3(2))
print(derivative(f3,2))
```

```python
import math
def f4(x):
    return math.cos(math.sqrt(math.e**x+1)/2)

print(f4(2))
print(derivative(f4,2))
```

```python
tmp1 = [1,2,3,4,5]
tmp2 = ['a','b','c','e','f']
tmp3 = [1,'a',2,'b',3,]

tmp_list = tmp1+tmp2+tmp3
print(tmp1+tmp2+tmp3,"\n")

for i in range(len(tmp_list)):
     print("요소의 인덱스 :",i,end='')
     print(" 해당하는 값 :",tmp_list[i],"\n")
```

```python
tmp_dic = {} 
tmp4 = ['ab','bc','cd','de','ef']
tmp5 = ["v",'w','x','y','z']

for i in range(len(tmp4)):
    if tmp4[i] not in tmp_dic:
        tmp_dic[tmp4[i]] = tmp5[i]
for key, value in tmp_dic.items():
        print(key,":",value)
```

```python
model_list = [['aa','bb','cc','dd'],['1a','2b','3c','4d'],['a1','b2','c3','d4']]
output = ''
for tmp_list in model_list:
    for index in range(len(tmp_list)):
        if index < len(tmp_list)-1:
            output += str(tmp_list[index])+','
        else:
            output += str(tmp_list[index])+'\n'
print(output,end='')
```
