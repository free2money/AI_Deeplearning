```python
def myTranspose(Matrix):
    (Matrix_row,Matrix_colom) = np.shape(Matrix)
    temp_matrix = np.zeros((Matrix_row,Matrix_colom))
    for row_list in Matrix:
        for row_index in range(len(row_list)):
            for colom_index in range(len(row_list)):
                if row_index == colom_index: 
                    temp_matrix[row_index][colom_index] = Matrix[row_index][colom_index]
                else:
                    temp_matrix[row_index][colom_index] = Matrix[colom_index][row_index]
    return temp_matrix
print("\n행렬 A의 전치행렬:\n",myTranspose(A))
```
